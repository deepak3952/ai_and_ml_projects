### b. DBSCAN Clustering as a preprocessing Technique

```python
df_events_location = df_events.loc[(df_events['latitude'] != 0) & (df_events['longitude'] != 0), ['latitude', 'longitude']]
```

```python
# scale the data
scaler = StandardScaler()
df_events_location_scaled = scaler.fit_transform(df_events_location)
```

```python
df_events_location_scaled = pd.DataFrame(df_events_location_scaled)
df_events_location_scaled.columns = ['F1', 'F2']
df_events_location_scaled.head()
```

```python
neigh = NearestNeighbors(n_neighbors=4)
nbrs = neigh.fit(df_events_location_scaled)
distances, indices = nbrs.kneighbors(df_events_location_scaled)
print(np.shape(distances))
neighbors = NearestNeighbors(n_neighbors=4)
neighbors_fit = neighbors.fit(df_events_location_scaled)
distances, indices = neighbors_fit.kneighbors(df_events_location_scaled)
```

```python
distances = np.sort(distances, axis=0)
plt.plot(distances)
plt.ylabel("Distance (Epsilon)")
plt.xlabel("Data points in ascending order of the distance")
plt.show()
```

```python
# create a range of min_samples values to test
min_samples_range = range(1, 11)
# perform DBSCAN clustering
# calculate the number of clusters for each min_samples value
num_clusters = []
for min_samples in min_samples_range:
    dbscan = dbscan = DBSCAN(eps=0.1, min_samples=min_samples)
    dbscan.fit(df_events_location)
    num_clusters.append(len(np.unique(dbscan.labels_)))

# plot the number of clusters against min_samples
plt.plot(min_samples_range, num_clusters)
plt.xlabel('min_samples')
plt.ylabel('Number of clusters')
plt.title('Elbow Method for DBSCAN')
plt.show()
```

```python
# perform DBSCAN clustering
kms_per_radian = 6371.0088
epsilon = 4.8/kms_per_radian
db = DBSCAN(eps=epsilon, min_samples=7, algorithm='ball_tree', metric='haversine').fit(df_events_location_scaled)
labels = db.labels_
```

```python
# visualize the clusters
plt.scatter(df_events_location['longitude'], df_events_location['latitude'], c=labels)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('DBSCAN Clustering')
plt.show()
```

### Kolmogorov–Smirnov (KS) statistic

```python
# obtain predicted probabilities for the test data
y_pred_prob = stacking_gender.predict_proba(X_test)

# separate the test data into positive and negative classes
pos_idx = np.where(y_test == 1)[0]
neg_idx = np.where(y_test == 0)[0]

# sort the predicted probabilities in descending order
pos_probs = y_pred_prob[pos_idx][:, 1]
neg_probs = y_pred_prob[neg_idx][:, 1]
pos_probs_sorted = np.sort(pos_probs)[::-1]
neg_probs_sorted = np.sort(neg_probs)[::-1]

min_len = min(len(pos_probs_sorted), len(neg_probs_sorted))
pos_cdf = np.cumsum(pos_probs_sorted[:min_len]) / np.sum(pos_probs_sorted[:min_len])
neg_cdf = np.cumsum(neg_probs_sorted[:min_len]) / np.sum(neg_probs_sorted[:min_len])

# compute the KS statistic as the maximum absolute difference between the CDFs
ks = np.max(np.abs(pos_cdf - neg_cdf))
ks
```

### The probability bands identified through the KS table for the top 3 and bottom 3 deciles on the train data set

```python

# get predicted probabilities for the positive class
y_proba_train = stacking_gender.predict_proba(X_train)[:, 1]

# calculate the false positive rate, true positive rate, and thresholds for the ROC curve
fpr, tpr, thresholds = roc_curve(y_train, y_proba_train)

# reset the index of y_train
y_train_reset = y_train.reset_index(drop=True)

# sort the predicted probabilities in descending order
idx = y_proba_train.argsort()[::-1]
y_proba_train_sorted = y_proba_train[idx]
y_train_sorted = y_train_reset[idx]

# calculate the number of samples and positive samples in each decile
n_samples = len(y_train_sorted)
n_pos_samples = y_train_sorted.sum()
n_neg_samples = n_samples - n_pos_samples
decile_size = n_samples // 10
pos_counts = np.cumsum(y_train_sorted)
neg_counts = np.arange(1, n_samples+1) - pos_counts

# calculate the CDFs for the positive and negative classes in each decile
pos_cdfs = pos_counts / n_pos_samples
neg_cdfs = neg_counts / n_neg_samples

# calculate the KS statistic and the threshold for the top 3 and bottom 3 deciles
ks_top = pos_cdfs[:decile_size].max() - neg_cdfs[:decile_size].max()
ks_bottom = pos_cdfs[-decile_size:].max() - neg_cdfs[-decile_size:].max()
thresh_top = y_proba_train_sorted[decile_size]
thresh_bottom = y_proba_train_sorted[-decile_size]

# calculate the probability bands for the top 3 and bottom 3 deciles
band_top = y_proba_train_sorted[:decile_size].min(), y_proba_train_sorted[:decile_size].max()
band_bottom = y_proba_train_sorted[-decile_size:].min(), y_proba_train_sorted[-decile_size:].max()

print(f"KS statistic for top 3 deciles: {ks_top:.3f}")
print(f"KS statistic for bottom 3 deciles: {ks_bottom:.3f}")
print(f"Probability band for top 3 deciles: {band_top}")
print(f"Probability band for bottom 3 deciles: {band_bottom}")
```
